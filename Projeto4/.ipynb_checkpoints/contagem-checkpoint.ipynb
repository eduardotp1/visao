## Introdução

Este projeto tem o intuito de compreender qual o background de determinadas imagens e tentar segmentar, assim, podemos obter apenas os objetos em movimentos de certa imagem, além disso, com a imagem segmentada, aplicamos um método de contagem, para saber quantas pessoas existem aproximadamente em um certa foto. Isto é relevante por poder mensurar de forma rápida quantas pessoas passam por certo local

### Técnica

Para chegar no resultado esperado, utilizamos o codebook, que está presente no Artigo 1, esta não é a única forma de obter a segmentação do fundo, possuindo, por exemplo, o método MOG, já presente no openCV, porém, o codebook dá uma melhor performance comparada ao MOG.

```python
#imports 
%matplotlib inline  
import pickle
import os
import glob
import cv2 as cv
import time
import numpy as np
import operations_helper as oph
import codeword
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
# figure size in inches optional
rcParams['figure.figsize'] = 15 ,12
```

```python
# setting important things

#todas as imagens devem ter a mesma resolução
folder_path = './people3'

data_path = os.path.join(folder_path,'*g')
files = glob.glob(data_path)

#list with all images of people in scene
data = []

print('Reading images...')
for f1 in files:
    img = cv.imread(f1)
    data.append(img)
print("Done")

print('Loading trainning')
with open('save3.pickle', 'rb') as handle:
    codebook = pickle.load(handle)
print('Done')

```

```python
#variables

e = 10 #epsolon aceito na variação rgb

#variaveis para range da iluminação
a = 0.5 
b = 1.2
```

```python
#função que determina se um pixel é background ou foreground
def get_foreground(img_bgr):
    img = img_bgr.copy()
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    for i in range(0,len(img)):
        for j in range(0, len(img[i])):
            pixel = img[i][j]
            brightness =  (pixel[0]**2 + pixel[1]**2 + pixel[2]**2)**(0.5)
            cb = codebook[i][j]
            for cw in cb:
                colorDist  = oph.get_color_distortion(pixel, cw.get_v())
                color_condition = float(colorDist) <=  float(e)
                ilow = cw.get_imax() * a
                ihi = min(b*cw.get_imax(),cw.get_imin()/a)
                brightness_condition  = (brightness >= ilow) and (brightness <= ihi)
                if(brightness_condition and color_condition):
                    img[i][j] = [0,0,0]
                else:
                    img[i][j] = [255,255,255]
    return img


```

## Decidindo , Erodindo e Dilatando

Primeiro usamos o codebook do treinamento para decidir o que é fundo e o que é elemento novo. Uma vez que a imagem resultante possui muito rúido de 'falsos' foregrounds o processo de erosão seguido de dilatação permite manter somente as zonas em que a densidade de pixels brancos são maiores.

```python
#Erodindo e Dilatando
def erode_dilate(imgin):
    img = get_foreground(imgin)

    imgGray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

    kernel = np.ones((5,5),np.uint8)
    
    erosion = cv.erode(imgGray,kernel,iterations = 1)
    
    dilate = cv.dilate(erosion,kernel,iterations = 1)


    
    return dilate
```

## Contagem de Pessoas

A contagem de pessoas se da por uma técnica simples e conveniente de indetificação de clusters na imagem. Basicamente, o algoritmo conta quantos "grupos" de pixels brancos conectados sobraram na imagem. È uma tecnica que nos da uma ideia boa da quantidade de pessoas, mas tem sua limitação, por exemplo, 2 pessoas andando juntas são 1 pessoa só, 1 carro é uma pessoa, qualquer elemento novo na cena com tamanho grande suficiente sera considerado uma pessoa a mais. Além de que, o processo de erosão e dilatação tem de ser bem cuidadoso para não quebrar uma pessoa em 2 grupos conectados.

```python
from skimage import measure

def count_connected_components(thresh):
    
    labels = measure.label(thresh, neighbors=8, background=0)
    mask = np.zeros(thresh.shape, dtype="uint8")
    
    number_of_spots = 0

    for label in np.unique(labels):

        if label == 0:
            continue


        labelMask = np.zeros(thresh.shape, dtype="uint8")
        labelMask[labels == label] = 255
        numPixels = cv.countNonZero(labelMask)

        if numPixels > 300:
            number_of_spots += 1
            mask = cv.add(mask, labelMask)
    return mask, number_of_spots
            


```
